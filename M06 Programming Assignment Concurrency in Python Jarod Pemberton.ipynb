```python
from datetime import date

# Get today's date
today = date.today()

# Write the date as a string to today.txt
with open("today.txt", "w") as file:
    file.write(str(today))
```

```python
# Read the date back into a string
with open("today.txt", "r") as file:
    today_string = file.read()

print(today_string)
```

```python
from datetime import datetime

# Convert the string back into a date object
parsed_date = datetime.strptime(today_string, "%Y-%m-%d").date()

print(parsed_date)
```

```python
import multiprocessing
import time
import random
from datetime import datetime

def show_time():
    # Sleep for a random time between 0 and 1 seconds
    time.sleep(random.random())
    # Print the current time
    print("Process time:", datetime.now())

if __name__ == "__main__":
    processes = []

    # Make three processes
    for _ in range(3):
        p = multiprocessing.Process(target=show_time)
        processes.append(p)
        p.start()

    for p in processes:
        p.join()

    print("All processes finished.")
```
